```python
fig,ax = plt.subplots(1,3,figsize=(9,3))
paths = sorted(Path('res').iterdir(), key=os.path.getmtime)
paths=paths[-300:]
# Method to change the contour data points
def animate(i):
    stage= paths[i]
    df = pd.read_csv(stage)
    NX= len(df.x.unique())
    NY= len(df.y.unique())

    x = df.x.unique()
    y = df.y.unique()
    ax[0].clear()
    ax[1].clear()
    ax[2].clear()
    
    X, Y = np.meshgrid(x, y)  
    z = df['psi'].to_numpy().reshape((NY, NX))
    cs = ax[0].contourf(X, Y, z,cmap=plt.cm.rainbow)
    #z = df['phi'].to_numpy().reshape((NY, NX))
   # cs = ax[1].contourf(X, Y, z,cmap=plt.cm.rainbow)
    ax[1].set_title(stage)
    z = df['T'].to_numpy().reshape((NY, NX))
    cs = ax[1].contourf(X, Y, z,cmap=plt.cm.rainbow)
    z = df['C'].to_numpy().reshape((NY, NX))
    cs = ax[2].contourf(X, Y, z,cmap=plt.cm.rainbow)
# Call animate method
ani = animation.FuncAnimation(fig, animate, len(paths), interval=5, blit=False)

# Display the plot
plt.show()
```

```python
def contur_plot(df):
    fig,ax = plt.subplots(1,3,figsize=(9,3))
    NX= len(df.x.unique())
    NY= len(df.y.unique())

    x = df.x.unique()
    y = df.y.unique()

    X, Y = np.meshgrid(x, y)  
    z = df['psi'].to_numpy().reshape((NY, NX))
    cs = ax[0].contourf(X, Y, z,cmap=plt.cm.rainbow)
    #z = df['phi'].to_numpy().reshape((NY, NX))
   # cs = ax[1].contourf(X, Y, z,cmap=plt.cm.rainbow)
    z = df['T'].to_numpy().reshape((NY, NX))
    cs = ax[1].contourf(X, Y, z,cmap=plt.cm.rainbow)
    z = df['C'].to_numpy().reshape((NY, NX))
    cs = ax[2].contourf(X, Y, z,cmap=plt.cm.rainbow)
    plt.show()
    
```
