```python
import  matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

import matplotlib.animation as animation
```

```python
def contur_plot(df):
    fig,ax = plt.subplots(1,3,figsize=(9,3))
    NX= len(df.x.unique())
    NY= len(df.y.unique())

    x = df.x.unique()
    y = df.y.unique()

    X, Y = np.meshgrid(x, y)  
    z = df['psi'].to_numpy().reshape((NY, NX))
    cs = ax[0].contourf(X, Y, z,cmap=plt.cm.rainbow)
    #z = df['phi'].to_numpy().reshape((NY, NX))
   # cs = ax[1].contourf(X, Y, z,cmap=plt.cm.rainbow)
    z = df['T'].to_numpy().reshape((NY, NX))
    cs = ax[1].contourf(X, Y, z,cmap=plt.cm.rainbow)
    z = df['C'].to_numpy().reshape((NY, NX))
    cs = ax[2].contourf(X, Y, z,cmap=plt.cm.rainbow)
    plt.show()
def contur_plot2(df):
    fig,ax = plt.subplots(figsize=(6,3))
    NX= len(df.x.unique())
    NY= len(df.y.unique())

    x = df.x.unique()
    y = df.y.unique()

    X, Y = np.meshgrid(x, y)  

    z = df['C'].to_numpy().reshape((NY, NX))
    cs = ax.contour(X, Y, z,colors='k')
    plt.savefig('1.png',dpi=600)
    plt.show()
    
    
```

```python
fig,ax = plt.subplots(1,3,figsize=(9,3))
paths = sorted(Path('res').iterdir(), key=os.path.getmtime)
paths=paths[:]
# Method to change the contour data points
def animate(i):
    stage= paths[i]
    df = pd.read_csv(stage)
    NX= len(df.x.unique())
    NY= len(df.y.unique())

    x = df.x.unique()
    y = df.y.unique()
    ax[0].clear()
    ax[1].clear()
    ax[2].clear()
    
    X, Y = np.meshgrid(x, y)  
    z = df['psi'].to_numpy().reshape((NY, NX))
    cs = ax[0].contourf(X, Y, z,cmap=plt.cm.rainbow)
    #z = df['phi'].to_numpy().reshape((NY, NX))
   # cs = ax[1].contourf(X, Y, z,cmap=plt.cm.rainbow)
    ax[1].set_title(stage)
    z = df['T'].to_numpy().reshape((NY, NX))
    cs = ax[1].contourf(X, Y, z,cmap=plt.cm.rainbow)
    z = df['C'].to_numpy().reshape((NY, NX))
    cs = ax[2].contourf(X, Y, z,cmap=plt.cm.rainbow)
# Call animate method
ani = animation.FuncAnimation(fig, animate, len(paths), interval=30, blit=False)

# Display the plot
plt.show()
```

```python
ani.save('1.avi')
```
