```python
import  matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import matplotlib.animation as animation
```

```python
df = pd.read_csv('res/stage_t=0.19531250000002784.dat')
```

```python
fig,ax = plt.subplots(4,figsize=(3,10))
NX= len(df.x.unique())
NY= len(df.y.unique())

x = df.x.unique()
y = df.y.unique()

X, Y = np.meshgrid(x, y)  
z = df['psi'].to_numpy().reshape((NY, NX))
cs = ax[0].contourf(X, Y, z,cmap=plt.cm.rainbow)
z = df['phi'].to_numpy().reshape((NY, NX))
cs = ax[1].contourf(X, Y, z,cmap=plt.cm.rainbow)
z = df['T'].to_numpy().reshape((NY, NX))
cs = ax[2].contourf(X, Y, z,cmap=plt.cm.rainbow)
z = df['C'].to_numpy().reshape((NY, NX))
cs = ax[3].contourf(X, Y, z,cmap=plt.cm.rainbow)
```
